```python
X = features
y = df['MEDV']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=1, test_size=0.3)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# baseline for both mse and mae
y_mean = y_train.mean()
y_pred_baseline = [y_mean] * len(y_train)

y_mse_baseline = mean_squared_error(y_train, y_pred_baseline)
y_mae_baseline = mean_absolute_error(y_train, y_pred_baseline)

print(f'baseline for MSE {y_mse_baseline} and baseline for MAE {y_mae_baseline}')
```

```python
from sklearn.linear_model import LinearRegression, Ridge
```

```python
model1= LinearRegression()
model2= Ridge()
```

```python
model1.fit(X_train, y_train)
```

```python
model2.fit(X_train, y_train)
```

```python
prediction = model1.predict(X_test)
model1_mse = mean_squared_error(y_test, prediction)
model1_mae = mean_absolute_error(y_test, prediction)
print(f'Model 1 MSE {model1_mse} and MAE {model1_mae}')
```

```python
prediction = model2.predict(X_test)
model2_mse = mean_squared_error(y_test, prediction)
model2_mae = mean_absolute_error(y_test, prediction)
print(f'Model 1 MSE {model2_mse} and MAE {model2_mae}')
```
