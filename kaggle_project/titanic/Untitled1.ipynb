```python
import pandas as pd
```

```python
data = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
data.head(20)
```

```python
def clean(data):
    data = data.drop(['Ticket', 'Cabin', 'Name', 'PassengerId'], axis=1)
    mean = data.Age.mean()
    data['Age'] = data['Age'].fillna(mean)
    data['Embarked'].fillna('U', inplace=True)
    return data
```

```python
data = clean(data)
test = clean(test)
```

```python
test.isnull().sum()
```

```python
data.Embarked.unique()
```

```python
test.Embarked.unique()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

cols = ['Sex', 'Embarked']
for col in cols:
    data[col] = le.fit_transform(data[col])
    test[col] = le.fit_transform(test[col])
    print(le.classes_)
    
data.head(10)
```

```python
y = data.Survived
features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']
X = data[features]
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import accuracy_score

rfmodel = RandomForestRegressor(random_state=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

rfmodel.fit(X_train, y_train)
rfprediction = rfmodel.predict(X_test)
rfmae = mean_absolute_error(y_test, rfprediction)
# rfas = accuracy_score(y_test, rfprediction)

print(f"Validation MAE for Random Forest Model: {rfmae}")
print("Validation accuracy_score for Random Forest Model: {:,.0f}".format(rfmae))
```

```python
from sklearn.linear_model import LogisticRegression

lrmodel = LogisticRegression()

lrmodel.fit(X_train, y_train)
lrprediction = lrmodel.predict(X_test)
lrmae = mean_absolute_error(y_test, lrprediction)
lras = accuracy_score(y_test, lrprediction)

print(f"Validation MAE for Logistic Regression Model: {lrmae}")
print("Validation accuracy_score for Logistic Regression Model: {:,.0f}".format(lras))
```
